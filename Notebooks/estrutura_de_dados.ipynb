**Tuplas**

```python
#AS tuplas usam parênteses como sintaxa
tp = ("Banana", "Maçã", 10, 50)
```

```python
#Retornando o primeiro elemento
tp[0]
```

```python
#Diferente das listas as tuplas são imutaveis, o que quer dizer que não podemos alterar os elementos
tp[0] = "Laranja"
```

```python
tp.count("Maçã")
```

```python
tp[0:2]
```

**Dicionarios**

```python
#Para criar um dicionário utilizamos as {}
dc = {"Maçã":20, "Banana":10, "Laranja":15, "Uva":5}#Dicionarios trabalham com o donceito chave e valor
```

```python
dc
```

```python
#Acessando o valor de um dicionarios atraves da chave
dc["Maçã"]
```

```python
#Atualizando o valor da maçã
dc["Maçã"] = 25
```

```python
dc
```

```python
#Retornando todas as chaves do dc
dc.keys()
```

```python
#Retornando os valores do dc
dc.values()
```

```python
#Verificando se já exite uma chave no dc e ncaso não exista inserir
dc.setdefault("Limão", 22)
```
